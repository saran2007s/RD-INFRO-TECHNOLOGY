**Handling Class Imbalance**

The dataset is highly imbalanced, with the majority of transactions being Legitimate (`Class = 0`) and only a small portion being fraudulent (`Class = 1`).

Understanding this imbalance is critical, as it can negatively impact the performance of standard classification models by biasing predictions toward the majority class.

We perform the following steps:
- **Inspect the distribution** of normal and fraudulent transactions.
- **Separate the data** into two subsets: `legit` and `fraud`, to allow further class-wise analysis.
- **Compute statistical summaries** (e.g., transaction amount distribution) for the two groups.


```python
# seperating the data for analysis
Legit = Dataset[Dataset.Class == 0]
Fraud = Dataset[Dataset.Class == 1]
```

```python
print(Legit.shape)
print(Fraud.shape)
```

```python
# statistical measures of the data
Legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Get mean values of features for each class (0 = Legit, 1 = fraud)
Dataset.groupby('Class').mean()
```

**Balancing the Dataset with Undersampling**


```python
# Sample the same number of legit transactions as fraud to balance the dataset
Legit_sample = Legit.sample(n=len(Fraud), random_state=1)  # added random_state for reproducibility
```

**Combining the Balanced Samples**

```python
# Combine sampled legit and all fraud transactions into a new balanced dataset
new_dataset = pd.concat([Legit_sample, Fraud], axis=0).sample(frac=1, random_state=1).reset_index(drop=True)
```

```python
# Display sample of the new dataset
print(new_dataset.shape)
print(new_dataset.head())
```

```python
# Count of each class to confirm balance
print(new_dataset['Class'].value_counts())
```

**Splitting the Data into Features and Target**

```python
features = new_dataset.drop(columns='Class', axis=1)
target = new_dataset['Class']
```

```python
# Check the shapes and preview
print("Features shape:", features.shape)
print("target shape:", target.shape)
print(features.head())
print(target.value_counts())
```

**Splitting the Dataset into Training and Testing Sets**

```python
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.2, stratify=target, random_state=2)
```

```python
# Check data splits
print(f"Original: {features.shape}, Train: {features_train.shape}, Test: {features_test.shape}")
```

**Model Training: Logistic Regression**

```python
# Initialize logistic regression model with default parameters
model = LogisticRegression(max_iter=5000, random_state=2)
```

```python
# Train the logistic regression model using the training data
model.fit(features_train, target_train)
```

**Model Evaluation**

**accuracy score**

```python
# Evaluate training accuracy
features_train_pred = model.predict(features_train)
train_acc = accuracy_score(target_train, features_train_pred)
print(f"Training Accuracy: {train_acc:.4f}")
```

```python
# Evaluate test accuracy
features_test_pred = model.predict(features_test)
test_acc = accuracy_score(target_test, features_test_pred)
print(f"Test Accuracy: {test_acc:.4f}")
```
